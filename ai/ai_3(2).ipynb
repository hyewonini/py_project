```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# 데이터 불러오기
df = pd.read_csv('Mall_Customers.csv')
# 기본 정보 확인
print(df.head()) # 데이터 상위 5행 확인
print(df.info()) # 데이터 타입과 결측치 확인
print(df.describe()) # 수치형 데이터의 요약 통계량
```

```python
# 결측값 확인
print(df.isnull().sum())

# 이상치 시각화 예제 - Age
df['Age'].hist(bins=10)
plt.xlabel('Age')
plt.show()
```

```python
print(df['Gender'].value_counts())    # Gender 변수 빈도수 확인

sns.countplot(x='Gender', data=df)
plt.show()
```

```python
# 결측값 확인
print(df.isnull().sum())

# 이상치 시각화 예제 - Annual Income (k$)
df['Annual Income (k$)'].hist(bins=20)
plt.xlabel('Annual Income (k$)')
plt.show()
```

```python
# 결측값 확인
print(df.isnull().sum())

# 이상치 시각화 예제 - Spending Score (1-100)
df['Spending Score (1-100)'].hist(bins=30)
plt.xlabel('Spending Score (1-100)')
plt.show()
```
